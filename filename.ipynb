```python
import os
import pandas as pd

# 디렉토리 경로 설정
directory = '/home/kng/kng/CRAWLER/2023년 경남 관광현황'

# 디렉토리 내의 모든 .xlsx 파일을 .csv 파일로 변환하고 기존 .xlsx 파일 삭제
for file_name in os.listdir(directory):
    if file_name.endswith('.xlsx'):
        # 파일 경로 설정
        file_path = os.path.join(directory, file_name)
        
        # 파일을 DataFrame으로 읽기
        df = pd.read_excel(file_path)
        
        # 새로운 파일 이름 설정
        new_name = file_name.replace('.xlsx', '.csv')
        new_path = os.path.join(directory, new_name)
        
        # DataFrame을 CSV 파일로 저장
        df.to_csv(new_path, index=False)
        print(f'Converted: {file_name} -> {new_name}')
        
        # 기존 .xlsx 파일 삭제
        os.remove(file_path)
        print(f'Deleted: {file_name}')
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 폰트 설정
font_path = '/home/kng/kng/CRAWLER/NanumGothic.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)

# 디렉토리 경로 설정
base_directory = '/home/kng/kng/CRAWLER'  # 기본 경로
years = ['2021년 경남 관광현황', '2022년 경남 관광현황', '2023년 경남 관광현황']

# 지역별 데이터를 저장할 사전
data_dict = {str(month).zfill(2): {} for month in range(1, 13)}  # 01, 02, ..., 12

# 각 년도 폴더에서 파일 처리
for year in years:
    directory = os.path.join(base_directory, year)  # 해당 년도의 경로
    
    for file_name in os.listdir(directory):
        if file_name.endswith('_merged.csv'):
            # 파일 경로 설정
            file_path = os.path.join(directory, file_name)
            
            # 파일을 DataFrame으로 읽기
            df = pd.read_csv(file_path)
            
            # 유동인구 수 정리
            df['유동인구 수'] = df['유동인구 수'].str.replace(',', '').str.replace(' 명', '').astype(int)
            
            # 월별 유동인구 수 집계
            for _, row in df.iterrows():
                month = row['구분'].split('년 ')[1].replace('월', '').zfill(2)  # "2021년 01월"에서 "01" 추출
                region = file_name.split('_')[0]  # 파일 이름에서 지역명 추출
                
                if month not in data_dict:
                    data_dict[month] = {}
                
                if region not in data_dict[month]:
                    data_dict[month][region] = []
                
                data_dict[month][region].append(row['유동인구 수'])

# 평균 유동인구 수 계산
average_data_dict = {region: [0]*4 for region in set(region for month in data_dict.values() for region in month.keys())}

# 계절별 인덱스: 0: 봄, 1: 여름, 2: 가을, 3: 겨울
season_months = {
    '봄': ['03', '04', '05'],
    '여름': ['06', '07', '08'],
    '가을': ['09', '10', '11'],
    '겨울': ['12', '01', '02']
}

for month, regions in data_dict.items():
    for region, values in regions.items():
        # 월에 따라 계절 인덱스 결정
        for season_index, (season, months) in enumerate(season_months.items()):
            if month in months:
                if len(values) > 0:
                    average_data_dict[region][season_index] += sum(values) / len(values)  # 평균 계산

# 각 지역별로 Min-Max 스케일링 적용
scaler = MinMaxScaler()
scaled_data_dict = {}
for region, values in average_data_dict.items():
    values = np.array(values).reshape(-1, 1)  # 스케일러에 넣기 위해 reshape
    scaled_values = scaler.fit_transform(values).flatten()  # 스케일링 후 평탄화
    scaled_data_dict[region] = scaled_values

# 계절별로 막대 그래프 그리기
for season_index, season in enumerate(season_months.keys()):
    plt.figure(figsize=(10, 6))
    regions = list(scaled_data_dict.keys())
    values = [scaled_data_dict[region][season_index] for region in regions]
    
    plt.bar(regions, values, color='skyblue')
    
    # 그래프 설정
    plt.title(f'2021~2023년 {season} 지역별 평균 유동인구 동향')
    plt.xlabel('지역')
    plt.ylabel('평균 유동인구 수 (스케일링)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # 그래프 보여주기
    plt.show()
```
