```python
import pandas as pd
import numpy as np

# 경상남도의 시 목록
cities = [
    "창원", "진주", "통영", "사천", "김해", "밀양", "거제", 
    "양산", "의령", "함안", "창녕", "고성", "남해", "하동", 
    "산청", "함양", "거창", "합천"
]

# 데이터 생성
data = {
    "지역": cities,
    "유동인구 수": np.random.randint(1000, 10000, size=len(cities)),  # 무작위 유동인구 수 (1000~10000)
}

# DataFrame 생성
df = pd.DataFrame(data)

# CSV 파일로 저장
output_file_path = '경상남도_유동인구.csv'
df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f'{output_file_path} 파일이 생성되었습니다.')
```
